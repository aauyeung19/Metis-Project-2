```python
#load movie stub and info pickles
movie_info = pickle.load(open('movie_info.pickle', 'rb'))
movie_stubs = pickle.load(open('movie_stubs.pickle', 'rb'))
```

```python
movie_info = pd.DataFrame(movie_info)
movie_info.set_index('title', inplace=True)
```

```python
movie_info.head()
```

```python
movie_stubs.head()
```

```python
movie_df = movie_stubs.merge(movie_info, left_index=True, right_index=True)

movie_df.drop_duplicates(inplace=True)
```

```python
#quick check to see if I have enough budget data
movie_df.budget.isnull().value_counts()
```

```python
#removing trailing number of theaters from opening column

def clean_openings(opening):
    """
    Opening String grabbed both money and the row below.
    $357,115,0074,662\n theaters
    uses regex to remove opening theaters and the string
    input: (str)
    return : (str) $###,###,###
    """
    try:
        return re.search('^[$][0-9]{1,3},[0-9]{3},[0-9]{3}', opening).group()
    except:
        return None
    
movie_df['opening'] = movie_df.opening.map(clean_openings)
```

```python
movie_df.filter(regex='Avengers', axis=0)
```

```python
def clean_money(money):
    """
    Convert Money from string to number. 
    
    input : (str) 
    return : (int)
    """
    try:
        return ''.join(money.split(','))[1:]
    except:
        return None
```

```python
movie_df.opening = movie_df.opening.map(clean_money)
movie_df.budget = movie_df.budget.map(clean_money)
movie_df.gross_dom = movie_df.gross_dom.map(clean_money)
movie_df.gross_inter = movie_df.gross_inter.map(clean_money)
movie_df.gross_world = movie_df.gross_world.map(clean_money)
```

```python
with open('clean_movie_df.pickle', 'wb') as to_write:
    pickle.dump(movie_df.head, to_write)
```
